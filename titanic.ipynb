```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Sarvadnya\\Downloads\\tested.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

```python
df['Fare'].describe()
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].median)
```

```python
df.info()
```

```python
df['Age'].unique()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'] = df['Fare'].fillna(df['Fare'].median())
```

```python
df.info()
```

```python
df.head()
```

```python
sns.countplot( x = 'Sex' , data = df)
```

**On the Titanic, the number of males was higher compared to females.**

```python
sns.countplot(data = df , x = 'Sex' , hue = 'Survived' , palette = 'Set2')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('No of People')
plt.legend(title = 'Survived' , labels = ['No' , 'Yes'])
plt.show()
```

**Most males did not survive, while most females did.**

```python
survivors = df[df['Survived'] == 1]


sns.countplot(data=survivors, x='Pclass', hue='Sex', palette='Set2')

plt.title('Number of Survivors by Pclass')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Number of Survivors')
plt.legend(title='Gender')
plt.show()
```

**Most surviving females were from 3rd class, fewer from 1st class, and the least from 2nd class**


```python
sns.countplot(data=df, x='Embarked', hue='Survived', palette='coolwarm')

plt.title('Survival Count by Embarked Location')
plt.xlabel('Port of Embarkation')
plt.ylabel('Number of People')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

**Most people from port S died, while ports C and Q had fewer deaths.**

```python
sns.countplot(data = df , x = 'SibSp' , hue = 'Survived' , palette = 'Greys')
plt.title('Survival Count by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard (SibSp)')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```

**Passengers with no siblings or spouses aboard had the highest death count, while survival dropped as the number increased.**

```python
sns.histplot(data=df, x='Fare', hue='Survived', multiple='stack', bins=30, palette='Set2')

plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare Paid')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```

**Most passengers paid low fares, and among them, more people died than survived.**

```python
sns.boxplot(data=df, x='Survived', y='Age', hue='Survived', palette='pastel', legend=False)

plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

**Most survivors were younger than those who did not survive, and children had a better chance of survival.**

```python
sns.pairplot(df, hue='Survived', vars=['Age', 'Fare', 'SibSp', 'Parch'])
```

**Survivors were generally younger, paid higher fares, and often had fewer family members on board.**

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')

plt.title('Fare vs Age by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

**Younger passengers who paid higher fares were more likely to survive.**
